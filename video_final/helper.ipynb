```python
import os
import shutil

# Set your video directory here
video_dir = 'C:/Users/USER/Downloads/Authentica---DeepFake-detection/Celeb-DF-2/Celeb-real/'

# Get all .mp4 files in the directory
all_files = sorted([f for f in os.listdir(video_dir) if f.endswith('.mp4')])

# List to keep track of original files already processed
processed_files = []

for file in all_files:
    original_path = os.path.join(video_dir, file)
    
    # If this file has already been processed, skip it
    if file in processed_files:
        continue

    # Check if this file is a copy (has pattern _1.mp4, _2.mp4, etc.)
    if any(file.startswith(orig_file.replace('.mp4', '')) and file != orig_file 
           for orig_file in processed_files):
        continue

    # Add this file to processed list
    processed_files.append(file)

    filename, ext = os.path.splitext(file)

    # Create 5 copies of the original file
    for i in range(1, 6):
        new_filename = f"{filename}_{i}{ext}"
        new_path = os.path.join(video_dir, new_filename)
        shutil.copyfile(original_path, new_path)
        print(f"Created: {new_filename}")

print("âœ… All files processed and copies created.")
```

```python
import os
import csv

def create_video_csv(input_dir, output_csv):
    # Get list of all .mp4 files
    video_files = [f for f in os.listdir(input_dir) if f.lower().endswith('.mp4')]

    # Open CSV file for writing (no header)
    with open(output_csv, mode='w', newline='') as file:
        writer = csv.writer(file)
        for video in video_files:
            writer.writerow([video, 'REAL'])

# Example usage
input_directory = 'C:/Users/USER/Downloads/Authentica---DeepFake-detection/Celeb-DF-v2/YouTube-real/'
output_csv_path = 'output3.csv'

create_video_csv(input_directory, output_csv_path)
```

```python
import csv
import random

def combine_and_shuffle_csv(csv_files, output_csv):
    combined_data = []

    # Read each CSV file and collect the rows
    for file in csv_files:
        with open(file, mode='r') as f:
            reader = csv.reader(f)
            combined_data.extend(list(reader))

    # Shuffle the combined list
    random.shuffle(combined_data)

    # Write to the output CSV without headers
    with open(output_csv, mode='w', newline='') as f:
        writer = csv.writer(f)
        writer.writerows(combined_data)

# Example usage
csv_files = ['file1.csv', 'file2.csv', 'file3.csv']
output_csv_path = 'Global_metadata.csv'

combine_and_shuffle_csv(csv_files, output_csv_path)
```
